```python
# HCF of a number 
def HCF(a,b):
    smaller = min(a,b)
    hcf = 0
    for i in range(1, smaller+1):
        if a%i == 0 and b%i == 0:
            hcf = i
    return hcf
HCF(45,60)
```

```python
# HCF using recursion
def HCF(a, b):
    if b == 0:
        return a
    return HCF(b, a%b)
HCF(45,60)
```

```python
# LCM of a number
def LCM(a,b):
    larger = max(a,b)
    lcm = 0
    while True:
        if larger % a == 0 and larger % b == 0:
            lcm = larger
            break
        larger += 1
    """for i in range(larger, (a*b)+1):
        if i % a == 0 and i % b == 0:
            lcm = i
            break"""
    return lcm
LCM(100,120)
```

```python
# finding the second largest number in a list
def secondLargestNumber(lst):
    first_max = max(lst[0], lst[1])
    sec_max = min(lst[0], lst[1])
    for i in range(len(lst)):
        if lst[i] >= first_max:
            if lst[i] > first_max:
                sec_max = first_max
                first_max = lst[i]
        else:
            if lst[i] > sec_max:
                sec_max = lst[i]
    return sec_max

lst = [4, 7, 2, 9, 4, 7, 3, 10, 12, 56, 45, 19, 44, 39, 14, 49, 30, 18]
result = secondLargestNumber(lst)
print("Second largest number :",result)
```

```python
# Pascal's Triangle (my Code)
def get_selections(n, r):
    if n-r < r:
        r = n - r
    if n <= 0 or r <= 0:
        return 1
    numerator = n
    for i in range(n-1, n-r, -1):
        numerator *= i
    denominator = 1
    for i in range(2, r+1):
        denominator *= i
    return int(numerator / denominator)

n = 5
for i in range(n):
    for j in range(n - i -1):
        print(" ", end = "")
    for k in range(i+1):
        print(get_selections(i, k), end = " ")
    print()
```

```python
# Pascal's Triangle (my code without using nCr concept)
n = 5
prev = [0, 1, 0]
results = [prev]
for i in range(1, n):
    curr = [0]
    for j in range(len(prev) - 1):
        curr.append(prev[j]+prev[j+1])
    curr.append(0)
    results.append(curr)
    prev = curr
    
for i in range(n):
    for j in range(n-i-1):
        print(" ", end = "")
    for val in results[i][1:-1]:
        print(val, end = " ")
    print()
```

```python
# Pascal's Triangle (GeeksForGeeks)
from math import factorial
 
# input n
n = 5
for i in range(n):
    for j in range(n-i+1):
 
        # for left spacing
        print(end=" ")
 
    for j in range(i+1):
 
        # nCr = n!/((n-r)!*r!)
        print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")
 
    # for new line
    print()
```
