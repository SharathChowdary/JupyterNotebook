```python
#dropping 'sl_no' column
data=data.drop('sl_no',axis=1)
```

```python
data
```

```python
#labelling the "status" column by converting Placed to 1 and Not Placed to 0
data.status=data.status.apply(lambda x : 1 if x=='Placed' else 0)
```

```python
data
```

```python
#checking the number of placed and not placed students in the dataset
data.status.value_counts()
```

```python
#dividing the data into two categories i.e. string datatypes and numberic datatypes
#string datatypes -> One Hot Encoding will be done
#Numberic datatypes -> Scaling will be done
cols_to_be_encoded=[]
cols_not_to_be_encoded=[]
for column in data.columns[:-1]:
    if data[column].dtype=='object':
        cols_to_be_encoded.append(column)
    else:
        cols_not_to_be_encoded.append(column)
cols_to_be_encoded,cols_not_to_be_encoded
```

```python
#importing OneHotEncoder method
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoded=encoder.fit_transform(data[cols_to_be_encoded]).toarray()
#OneHotEncoder() will returns an object and hence it is converted to array using toarray()
```

```python
#importing MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler() #converts values between 0 and 1
scaled=scaler.fit_transform(data[cols_not_to_be_encoded])
scaled=pd.DataFrame(scaled,columns=['col1','col2','col3','col4','col5']) 
#Here sample column names are assigned so that there is no confusion while joining the DataFrames later. Otherwise error is thrown.
scaled
```

```python
#joing two dataframes
x=pd.DataFrame(encoded).join(scaled)
```

```python
x
```

```python
y=data.status
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the data into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3,stratify=y)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
#importing LogisticRegression model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
```

```python
#training the model
model.fit(x_train,y_train)
```

```python
#checking the accuracy of the model
model.score(x_test,y_test)
```

```python
data.columns
```

```python
def predict(input_list):
    labels=["Not Placed","Placed"]
    num_cols=[]
    string_cols=[]
    for val in input_list:
        if type(val)==str:
            string_cols.append(val)
        else:
            num_cols.append(val)
    encoded=encoder.transform([string_cols]).toarray() #encoding string values
    scaled=scaler.transform([num_cols]) #scaling numerical values
    x=np.append(encoded,scaled) #appending two numpy arrays
    output=model.predict([x]) #predicting the output
    return labels[output[0]]   
```

```python
input_list=['M',61,'Others',68,"Others","Science",64,"Sci&Tech","No",65,"Mkt&Fin",67]
predict(input_list)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#confusion matrix
labels=["Not Placed","Placed"]
cm=confusion_matrix(y_test,y_predicted)
sns.heatmap(cm,annot=True,xticklabels=labels,yticklabels=labels)
plt.xlabel("Truth")
plt.ylabel("Predicton")
plt.show()
```

```python
#classification report
print(classification_report(y_test,y_predicted))
```

# Hyper parameter tuning

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
```

```python
models = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(),
        'params': {
            'C': [1,5,10]
        }
    },
    'naive_bayes_gaussian': {
        'model': GaussianNB(),
        'params': {}
    },
    'naive_bayes_multinomial': {
        'model': MultinomialNB(),
        'params': {}
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini','entropy'],
            
        }
    }     
}
```

```python
results=[]
for model_name,details in models.items():
    model=GridSearchCV(details['model'],details['params'],return_train_score=False,cv=3)
    model.fit(x,y)
    results.append({
        'Model':model_name,
        'Best score':model.best_score_,
        'Best parameters':model.best_params_
    })
results_df=pd.DataFrame(results)
results_df
```
