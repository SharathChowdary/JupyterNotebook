```python
#Eliminating columns which don't have impact on Churn
impact_columns=['SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod','Churn']
```

```python
data1=data[impact_columns]
```

```python
data1
```

```python
for column in data1:
    print(f"{column} : {data1[column].unique()}")
```

```python
#both 'No internet service' and 'No' indicates no hence replaced with 'No'
data1=data1.replace('No internet service','No')
```

```python
for column in data1:
    print(f"{column} : {data1[column].unique()}")
```

```python
#scaling the values
cols_to_be_scaled=['tenure']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
data1[cols_to_be_scaled]=scaler.fit_transform(data1[cols_to_be_scaled])
```

```python
data1.head()
```

```python
#one hot encoding
data2=pd.get_dummies(data1,drop_first=True)
```

```python
data2
```

```python
x=data2.drop('Churn_Yes',axis=1)
x
```

```python
y=data2.Churn_Yes
y
```

**From the above scores we can observe that Linear Regression is performing better**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
model=LogisticRegression(C=5)
```

```python
#splitting the data into training and testing samples
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
#training a model
model.fit(x_train,y_train)
```

```python
#checking the accuracy of the model
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
cm=confusion_matrix(y_predicted,y_test)
sns.heatmap(cm,annot=True)
plt.xlabel("Truth")
plt.ylabel("Predicted")
plt.show()
```

```python
print(classification_report(y_predicted,y_test))
```
