```python
#importing libraries
import tensorflow as tf
import pathlib
import matplotlib.pyplot as plt
import cv2 as cv
import PIL
```

```python
path=pathlib.Path("/kaggle/input/face-mask-detection/Dataset")
```

```python
with_mask=list(path.glob("with_mask/*"))
without_mask=list(path.glob("without_mask/*"))
mask_weared_incorrect=list(path.glob("mask_weared_incorrect/*"))
```

```python
PIL.Image.open(with_mask[0])
```

```python
PIL.Image.open(without_mask[0])
```

```python
PIL.Image.open(mask_weared_incorrect[0])
```

```python
label_names=['With mask','Without mask','Mask weared incorrectly']
```

```python
dct={
    "with_mask":with_mask,
    "without_mask":without_mask,
    "mask_weared_incorrect":mask_weared_incorrect
}
```

```python
features=[]
labels=[]
```

```python
#creating x and y variables
for label_name,paths in dct.items():
    for path in paths:
        img=PIL.Image.open(path)
        features.append(np.array(img))
        labels.append(label_names.index(label_name))
```

```python
len(features)
```

```python
len(labels)
```

```python
x=np.array(features)
y=np.array(labels)
```

```python
np.unique(labels)
```

```python
#splitting the data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3,stratify=y)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
#initializing the model
model=tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=10,kernel_size=(5,5),activation='relu',input_shape=(128,128,3)),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(filters=20,kernel_size=(5,5),activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(filters=30,kernel_size=(5,5),activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(15,activation='relu'),
    tf.keras.layers.Dense(10,activation='relu'),
    tf.keras.layers.Dense(3,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
             loss=tf.losses.sparse_categorical_crossentropy,
             metrics=['accuracy'])
```

```python
#training the model
model.fit(x_train,y_train,epochs=15)
```

```python
#checking the accuracy of the model
model.evaluate(x_test,y_test)
```

```python
predicted=model.predict(x_test)
```

```python
#saving the model
model.save('face_mask_detection.h5')
```

```python
import tensorflow as tf
import numpy as np
import cv2 as cv
```

```python
#loading the model
model=tf.keras.models.load_model('face_mask_detection_model.h5',compile=False)
```

```python
model.summary()
```

```python
label_names=['With mask','Without mask','Mask weared incorrectly']
```

```python
def predict():
    #importing face detection haar cascade
    haar_cascade=cv.CascadeClassifier("D:\programming languages\OpenCV\HaarCascades\haarcascade_frontalface_default.xml")
    capture=cv.VideoCapture(0)
    while True:
        isTrue,frame=capture.read()
        if isTrue:
            gray=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
            faces=haar_cascade.detectMultiScale(gray,1.1,10)
            roi=frame #initializing roi with initial frame because if no face detected then error is thrown
            X,Y,W,H=0,0,0,0 #initializing with zeros for the above reason
            if len(faces)==0:
                cv.putText(frame,"Keep your face straight to camera",(10,30),cv.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
            else:
                for x,y,w,h in faces:
                    X,Y,W,H=x,y,w,h
                    roi=frame[y:y+h,x:x+w]
                    resized=cv.resize(roi,(128,128))
                    converted=cv.cvtColor(resized,cv.COLOR_BGR2RGB) #cv reads image in BGR and hence is converted to RBG
                    img=np.array(converted) #converting image into an array
                    img=img[np.newaxis,...] #(128,128,3) --> (1,128,128,3)
                    predicted=model.predict(img)
                    confidence=round(np.max(predicted)*100,2)
                    label=np.argmax(predicted)
                    if label==0:
                        color=(0,255,0) #if mask is there then green
                    elif label==1:
                        color=(0,0,255) #if mask is not there then red
                    else:
                        color=(0,255,255) #otherwise yellow
                    cv.putText(frame,label_names[label],(X,Y-30),cv.FONT_HERSHEY_COMPLEX,0.5,color,2)
                    cv.putText(frame,"Confidence : "+str(confidence),(X,Y-10),cv.FONT_HERSHEY_COMPLEX,0.5,color,2)
                    cv.rectangle(frame,(X,Y),(X+W,Y+H),color,2)
            cv.imshow("Face",frame)
            if cv.waitKey(1) & 0xFF==ord('q'):
                break
        else: 
            break
    capture.release()
    cv.destroyAllWindows()
```

```python
predict()
```
