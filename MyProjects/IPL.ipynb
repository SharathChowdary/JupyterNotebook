```python
data2=data1[data1.venue.isin(stadiums)]
data2
```

```python
data2.date=pd.to_datetime(data2.date)
```

```python
data2.date.dtype
```

```python
data3=data2.drop(['mid','batsman','bowler','striker','non-striker'],axis=1)
data3
```

```python
data4=pd.get_dummies(data3,columns=['venue','bat_team','bowl_team'],drop_first=True)
data4
```

```python
data4.reset_index(inplace=True)
data4
```

```python
data4.drop('index',axis=1,inplace=True)
data4
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
columns_to_be_scaled=['runs','wickets','overs','total']
scaled_columns=scaler.fit_transform(data3[columns_to_be_scaled])
scaled_df=pd.DataFrame(scaled_columns,columns=columns_to_be_scaled)
scaled_df
```

```python
data4.drop(columns_to_be_scaled,axis=1,inplace=True)
data4
```

```python
data5=pd.concat([data4,scaled_df],axis=1)
data5
```

```python
data3=data5
```

```python
training_data=data3[data3.date.dt.year<2016].drop('date',axis=1)
training_data
```

```python
testing_data=data3[data3.date.dt.year>=2016].drop('date',axis=1)
testing_data
```

```python
x_train=training_data.drop(['total'],axis=1)
x_train
```

```python
x_test=testing_data.drop('total',axis=1)
x_test
```

```python
y_train=training_data.total
y_train
```

```python
y_test=testing_data.total
y_test
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score
```

```python
ridge = Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
ridge_regressor = RandomizedSearchCV(ridge,parameters,cv=10,scoring='neg_mean_squared_error')
ridge_regressor.fit(x_train,y_train)
```

```python
ridge_regressor.score(x_test,y_test)
```

```python
prediction_r = ridge_regressor.predict(x_test)
r2_score(y_test,prediction_r)
```
